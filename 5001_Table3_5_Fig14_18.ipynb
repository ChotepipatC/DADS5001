```python
for f in ['ประเภทองค์กร','ประเภทหน่วยบริการสุขภาพ','สังกัด','แผนก/กรม']:
  print(f"{f}: {health_office_clean[f].unique()}")
```

```python
def hos_cate(row):
  if row['ประเภทองค์กร'] == 'รัฐบาล':
    if row['ประเภทหน่วยบริการสุขภาพ'] == 'โรงพยาบาลศูนย์':
      return 'โรงพยาบาลศูนย์'
    elif row['สังกัด'] == 'กระทรวงการอุดมศึกษา วิทยาศาสตร์ วิจัยและนวัตกรรม':
      return 'โรงพยาบาลมหาวิทยาลัย'
    return 'สถานพยาบาลรัฐบาลอื่นๆ'
  return 'สถานพยาบาลเอกชน'
```

```python
health_office_clean['hospital_cate'] = health_office_clean.apply(hos_cate, axis=1)

health_office_clean.sample(10)
```

```python
h_cate_df = health_office_clean[['รหัส 5 หลัก','hospital_cate']].copy()

combined_clean = pd.merge(combined_clean, h_cate_df,
                       left_on='hcode',
                       right_on='รหัส 5 หลัก',
                       how='left')
```

```python
# Sum N_service by Province

select_office = list(health_office_clean['รหัส 5 หลัก'].unique())
pro_f = combined_clean[combined_clean['hcode'].isin(select_office)]
province_service = pro_f.pivot_table(index='prov', values='n_service',aggfunc='sum').sort_values(by='n_service', ascending=False)
province_service.reset_index(inplace=True)
province_service.columns = ['province', 'sum_n_service']

pro_f_start = pro_f[pro_f['month_no']==202210]
province_service_start = pro_f_start.pivot_table(index='prov', values='n_service',aggfunc='sum').sort_values(by='n_service', ascending=False)
province_service_start.reset_index(inplace=True)
province_service_start.columns = ['province', 'sum_n_service_start']

pro_f_end = pro_f[pro_f['month_no']==202409]
province_service_end = pro_f_end.pivot_table(index='prov', values='n_service',aggfunc='sum').sort_values(by='n_service', ascending=False)
province_service_end.reset_index(inplace=True)
province_service_end.columns = ['province', 'sum_n_service_end']

# hospital category
province_hos_cate = pro_f.pivot_table(index='prov',columns='hospital_cate', values='n_service',aggfunc='sum')
province_hos_cate.reset_index(inplace=True)
province_hos_cate = province_hos_cate.rename(columns={'prov':'province',
 'สถานพยาบาลรัฐบาลอื่นๆ':'oth_gov_service',
 'สถานพยาบาลเอกชน':'private_service',
 'โรงพยาบาลมหาวิทยาลัย':'uni_hos_service',
 'โรงพยาบาลศูนย์':'central_hos_service'})
province_hos_cate = province_hos_cate[['province','central_hos_service','uni_hos_service','oth_gov_service','private_service']]


# Sum UC bed by Province
select_office = list(combined_clean['hcode'].unique())
office_f = health_office_clean[health_office_clean['รหัส 5 หลัก'].isin(select_office)]
province_ucbed = office_f.pivot_table(index='จังหวัด', values='เตียงที่ใช้จริง',aggfunc='sum') #.sort_values(by='เตียงที่ใช้จริง', ascending=False)
province_ucbed.reset_index(inplace=True)
province_ucbed.columns = ['province', 'sum_uc_bed']

# Sum population by Province
province_pop = popu[popu['yyyymm'] == max(popu['yyyymm'])][['province','total_pop']].copy()
province_pop_e = popu[popu['yyyymm'] == max(popu['yyyymm'])][['province','total_pop']].copy()
province_pop_e.columns = ['province','total_pop_end']
province_pop_s = popu[popu['yyyymm'] == min(popu['yyyymm'])][['province','total_pop']].copy()
province_pop_s.columns = ['province','total_pop_start']

# Sum all bed by Province
province_allbed = health_office_clean.pivot_table(index='จังหวัด', values='เตียงที่ใช้จริง',aggfunc='sum') #.sort_values(by='เตียงที่ใช้จริง', ascending=False)
province_allbed.reset_index(inplace=True)
province_allbed.columns = ['province', 'sum_all_bed']

#region
reg_5reg = reg[['ADM1_TH','REGION4']].copy()
reg_5reg.columns = ['province','region']

# Combined
bed_usage = pd.merge(province_service, province_ucbed, on='province',how='inner')
bed_usage = pd.merge(bed_usage, province_pop, on='province',how='inner')
bed_usage = pd.merge(bed_usage, province_pop_s, on='province',how='inner')
bed_usage = pd.merge(bed_usage, province_pop_e, on='province',how='inner')
bed_usage = pd.merge(bed_usage, province_hos_cate, on='province',how='inner')
bed_usage = pd.merge(bed_usage, province_allbed, on='province',how='inner')
bed_usage = pd.merge(bed_usage, reg_5reg, on='province',how='inner')
bed_usage = pd.merge(bed_usage, province_service_start, on='province',how='inner')
bed_usage = pd.merge(bed_usage, province_service_end, on='province',how='inner')

# Rate cal
bed_usage['bed_usage_rate'] = bed_usage['sum_n_service']/bed_usage['sum_uc_bed']
bed_usage['pop_per_bed'] = bed_usage['total_pop']/bed_usage['sum_all_bed']
bed_usage['growth'] = ((bed_usage['sum_n_service_end']/bed_usage['sum_n_service_start'])-1)*100
bed_usage['pop_growth'] = ((bed_usage['total_pop_end']/bed_usage['total_pop_start'])-1)*100

#sort
bed_usage = bed_usage.sort_values(by='bed_usage_rate', ascending=False)

#province_service
#province_ucbed
bed_usage
```

```python
select_office = list(health_office_clean['รหัส 5 หลัก'].unique())
pro_f = combined_clean[combined_clean['hcode'].isin(select_office)]
pdx_service = pro_f.pivot_table(index=['prov','pdx','pdx_name'], values='n_service'
                                ,aggfunc='sum').sort_values(by=['prov','n_service']
                                                            , ascending=False)
pdx_service.reset_index(inplace=True)
pdx_service.columns = ['province', 'pdx', 'pdx_name', 'sum_n_service']

reg_5reg = reg[['ADM1_TH','REGION4']].copy()
reg_5reg.columns = ['province','region']

pdx_mas = pdx_master[['pdx','Thai']]

pdx_service = pd.merge(pdx_service,reg_5reg,on='province',how='left')
pdx_service = pd.merge(pdx_service,pdx_mas,on='pdx',how='left')

pdx_service = pd.DataFrame( pdx_service[['region','pdx','pdx_name','Thai','sum_n_service']].\
              groupby(by=['region','pdx','pdx_name','Thai']).sum().reset_index() )

# Step 1: Rank within each region

pdx_service['rank'] = pdx_service.groupby('region')['sum_n_service'] \
                                 .rank(method='first', ascending=False) \
                                 .astype("Int32")

# Step 2: Keep only Top 5
top5 = pdx_service[pdx_service['rank'] <= 5].copy()

# Step 3: Pivot so each region is a column of ranks
pivot = top5.pivot_table(index=['pdx_name','Thai'],
                         columns='region',
                         values='rank',
                         aggfunc='first')

# Step 4: Calculate total sum_n_service across all regions
total_service = pdx_service.groupby(['pdx_name','Thai'])['sum_n_service'] \
                           .sum().rename("sum_n_service_total")

# Step 5: Merge total into pivot
result = pivot.merge(total_service, on=['pdx_name','Thai'])

# Step 6: Reset index and sort
#result = result.reset_index().fillna("")
result = result.reset_index()
result = result.sort_values(by="sum_n_service_total", ascending=False)

# Define the desired column order
region_cols = list(pivot.columns)
ordered_cols = ['pdx_name','Thai','sum_n_service_total'] + region_cols

# Reorder columns
result = result[ordered_cols]

# Sort rows by sum_n_service_total (descending for top diseases)
result = result.sort_values(by='sum_n_service_total', ascending=False).reset_index(drop=True)

result
```

```python
import matplotlib.pyplot as plt
import math

# --- helper: round up nicely ---
def round_up_nice(value):
    if value == 0:
        return 0
    magnitude = 10 ** (len(str(int(value))) - 1)
    step = magnitude // 2 if value/magnitude < 2 else magnitude
    return math.ceil(value / step) * step

# # --- determine axis limits ---
# max_x = round_up_nice(bed_usage['pop_per_bed'].max())
# max_y = round_up_nice(bed_usage['bed_usage_rate'].max())

max_x = math.ceil( max(bed_usage['pop_per_bed'])/100 )*100
max_y = math.ceil( max(bed_usage['bed_usage_rate'])/100 )*100

# --- region colors (manual mapping) ---
region_colors = {'ภาคเหนือ': 'purple',
                  'ภาคตะวันออกเฉียงเหนือ': 'red',
                 'ภาคกลาง': 'blue',
                 'ภาคตะวันออก': 'green',
                 'ภาคใต้': 'orange'
}

fig, ax = plt.subplots(figsize=(6,4), dpi=300)
ax.set_xlim(0, max_x)
ax.set_ylim(0, max_y)

# --- scatter plot ---
for region, color in region_colors.items():
    subset = bed_usage[bed_usage['region'] == region]
    ax.scatter(
        subset['pop_per_bed'],
        subset['bed_usage_rate'],
        s=subset['sum_n_service']/2500,
        color=color,
        alpha=0.5,
        label=region
    )

ax.set_xlabel("ประชากรต่อเตียง (คนต่อเตียง)")
ax.set_ylabel("อัตราการใช้เตียง (ครั้งต่อเตียง)")
ax.legend(title="ภูมิภาค", loc="upper center",
    bbox_to_anchor=(0.5, -0.20),   # center below the plot
    ncol=5,     # all labels in 1 row
    frameon=True                  )

# --- annotate provinces with max values ---
# 1) max pop_per_bed
prov_pop = bed_usage.loc[bed_usage['pop_per_bed'].idxmax()]
ax.annotate(prov_pop['province'],
            (prov_pop['pop_per_bed'], prov_pop['bed_usage_rate']),
            xytext=(5,5), textcoords='offset points',
            fontsize=8, color='blue')

# 2) max bed_usage_rate
prov_rate = bed_usage.loc[bed_usage['bed_usage_rate'].idxmax()]
ax.annotate(prov_rate['province'],
            (prov_rate['pop_per_bed'], prov_rate['bed_usage_rate']),
            xytext=(5,-10), textcoords='offset points',
            fontsize=8, color='red')

# 3) max sum_n_service
prov_serv = bed_usage.loc[bed_usage['sum_n_service'].idxmax()]
ax.annotate(prov_serv['province'],
            (prov_serv['pop_per_bed'], prov_serv['bed_usage_rate']),
            xytext=(-40,5), textcoords='offset points',
            fontsize=8, color='green')

plt.show()
```
